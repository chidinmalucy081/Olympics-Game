```python
#DATA CLEANING
130 rows × 7 columns has missing values



teams[teams.isnull().any(axis=1)]
```

```python
#DECIDED TO DROP THE MISSING DATA

teams = teams.dropna()
```

```python
teams
```

```python
#SPLIT DATA~ reason is to train our model evaluate how well the model is
#because this data is a time series data some rows comes 
#before some rows we wanna take the last 2years of data we have 2012 & 2016
#and put them in TESTING DATA SET & previous year and put in the TRAINING DATA SET



train = teams[teams["year"] < 2012].copy()
test = teams[teams["year"] >= 2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
#TRAIN MODEL

from sklearn.linear_model import LinearRegression
Reg = LinearRegression()
```

```python
#DEFINE PREDICTORS & TARGETS (we are using predictors to predict target)
predictors = ["athletes", "prev_medals"]
target =  ["medals"]
```

```python
#fit method

Reg.fit(train[predictors], train["medals"])
```

```python
predictions = Reg.predict(test[predictors])
```

```python
test["predictions"] = predictions
```

```python
test
```

```python
#this will replace all predictions that are less than 0 to 0
test.loc[test["predictions"] < 0, "predictions"] = 0
```

```python
#this will make it around figure
test["predictions"] = test["predictions"].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test["medals"], test["predictions"])
```

```python
# make sue your error is below the standard deviation.
error
```
